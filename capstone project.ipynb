```python
### missing data percentage greater than 25 in any column can be removed
```

```python
total_data = df.isnull().sum().sort_values(ascending = False)
percentage = (df.isnull().sum()/ df.isnull().count() * 100).sort_values(ascending = False)
df_percentage = pd.concat([total_data,percentage] , axis = 1 , keys = ['Total' , 'percent'] )
df_percentage.head(20)
```

```python
### Dropping the missing columns more than 25 percentage

for i in df.columns:
    if df_percentage.loc[i,'percent'] >= 25:
        df=df.drop(i,axis=1)
```

```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

### CHI2 TEST

```python
from sklearn.feature_selection import chi2
f_p_values = chi2(x1,y)
```

```python
f_p_values
```

```python
## good features should have high f value and very less p value(<0.05)

p_values = pd.Series(f_p_values[1])
p_values.index = x1.columns
p_values.nsmallest(15)
```

```python
from sklearn.feature_selection import SelectKBest
features = SelectKBest(score_func = chi2 , k=20)
ordered_feature=features.fit(x1,y)
```

```python
dfscores=pd.DataFrame(ordered_feature.scores_,columns=["Score"])
dfscores.index = x1.columns
```

### Train  Test  Split 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

### RESAMPLING USING SMOTETomek

```python
from imblearn.combine import SMOTETomek
```

```python
os=SMOTETomek(0.75)
x_res,y_res=os.fit_sample(x_train,y_train)
from collections import Counter
print('Original dataset shape {}'.format(Counter(y_train)))
print('Resampled dataset shape {}'.format(Counter(y_res)))
```

```python
x_res.shape,y_res.shape 
```

### Model Building with balanced dataset

```python
from sklearn.model_selection import cross_val_score 
from sklearn.metrics import roc_auc_score , roc_curve , auc , accuracy_score ,classification_report,confusion_matrix ,recall_score,precision_score,f1_score
```

### RandomForestClassifier 

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators= 100 , criterion= 'gini')
rfc.fit(x_res , y_res)
```

```python
y_pred = rfc.predict(x_test)
print(confusion_matrix(y_test , y_pred))
```

```python
print('accuracy_score is :{}'.format(accuracy_score(y_test , y_pred)))
print('\nroc_auc_score is :{}'.format(roc_auc_score(y_test,y_pred)))

print('precision_score is :{}'.format(precision_score(y_test,y_pred)))
print('recall_score is :{}'.format(recall_score(y_test,y_pred)))
print('f1_score is :{}'.format(f1_score(y_test,y_pred)))

print('\n' ,classification_report(y_test,y_pred))
```

```python
sns.heatmap(confusion_matrix(y_test , y_pred ) , annot = True,fmt=".1f",linecolor='white',linewidths=1)
plt.ylabel('actual label')
plt.xlabel('predicted label')
```

## Cross Validation
